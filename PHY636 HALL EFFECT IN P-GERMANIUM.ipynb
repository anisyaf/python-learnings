```python
import matplotlib.pyplot as plt
import numpy as np

control_current = [-30, -25, -20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30]
hall_voltage = [-5.48, -5.46, -5.43, -5.41, -5.39, -5.36, -5.34, -5.31, -5.27, -5.24, -5.21, -5.17, -5.13]

plt.plot(control_current, hall_voltage)
plt.plot(control_current, hall_voltage, color = 'green')
plt.scatter(control_current, hall_voltage, marker = 'o', s=50)
plt.xlabel('Current, I (mA)')
plt.ylabel('Hall Voltage, VH (V)')
plt.title('Hall voltage as a function of current')
plt.figtext(0.5,0.60, 'B=250 mT, room temperature', ha='right')
plt.grid(False)
plt.savefig('my_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

magnetic_flux = [20.1, 40.3, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300]
r_ro = [-0.6264, -0.6264, -0.6258, -0.6247, -0.6244, -0.6234, -0.6233, -0.6222, -0.6218, -0.6209, -0.6198, -0.6191, -0.6178, -0.6173, -0.6162]

plt.plot(magnetic_flux, r_ro)
plt.plot(magnetic_flux, r_ro, color = 'green')
plt.scatter(magnetic_flux, r_ro, marker = 'o', s=50)
plt.xlabel('Magnetic Flux, B, (mT)')
plt.ylabel('(R-Ro)/Ro)')
plt.title('Change of resistance as a function of magnetic field')
plt.grid(False)
plt.savefig('graph2.png')
plt.show()
```
